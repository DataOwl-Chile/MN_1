### 8.2 Paseos Aleatorios

Hemos visto ya cómo utilizar una función densidad de probabilidad para dar peso a ciertos rangos de valores por sobre otros en el cálculo de integrales. Esta elección la hicimos sólo viendo la forma de la función que deseamos integrar, lo cual matemáticamente no se sostiene mucho. Sin embargo, es posible formalizar este método a través del **Teorema de Bayes**.

Dado un espacio muestral $\Omega$ (discreto o continuo) y dos eventos $A,B\subseteq\Omega$, se puede calcular la probabilidad de que ocurra $A$ si ya sabemos que $B$ ha ocurrido. Esto define la **probabilidad condicional** del evento $A$, condicional a $B$, y se escribe como

$$\mathbb{P}\left[A|B\right]=\frac{\mathbb{P}[A\cap B]}{\mathbb{P}[B]}$$

y se lee como "la probabilidad de $A$ dado $B$". Por supuesto, esto adquiere relevancia cuando la ocurrencia de $B$ influye en la probabilidad de ocurrencia de $A$, ya que si $A$ es independiente de $B$, $\mathbb{P}[A\cap B]=\mathbb{P}[A]\cdot\mathbb{P}[B]$, y por lo tanto $\mathbb{P}\left[A|B\right]=\mathbb{P}\left[A\right]$.

El Teorema de Bayes establece que

$$\mathbb{P}\left[A|B\right]=\frac{\mathbb{P}\left[B|A\right]\cdot\mathbb{P}[A]}{\mathbb{P}[B]}$$

es decir, que la probabilidad de que ocurra $A$ dado que ocurrió $B$ es proporcional a la probabilidad de que ocurra $B$ dado que ocurrió $A$. Esta fórmula tiene múltiples interpretaciones, según la lectura que se dé a los eventos.

Otra forma que adquiere las probabilidades condicionales está dada por la función densidad de una variable aleatoria $X$ y una variable aleatoria $Y$:

$$f_{X|Y=y}(x)=\frac{f_{X,Y}(x,y)}{f_Y(y)}$$

donde $f_{X,Y}$ es la densidad conjunta de las variables aleatorias $X$ e $Y$. Esta última se interpreta como la densidad que tienen los vectores $(X,Y)$ en el espacio muestral de todas las posibles combinaciones entre ambos espacios muestrales, $\Omega_X\times\Omega_Y$. Luego, el Teorema de Bayes dice que

$$f_{X|Y=y}(x)=\frac{f_{Y|X=x}(y)f_X(x)}{f_Y(y)}$$

Por lo general, nuestra noción de probabilidad está familiarizada con la probabilidad **frecuentista**, donde las probabilidades se calculan basándonos en una distribución conocida por la naturaleza del problema o bien por contar con muchos datos que respaldan este conocimiento. Cuando los datos son insuficientes como para dar cuenta de una distribución claramente definida pero aun así tenemos una creencia sobre cuál podría ser, se utiliza la probabilidad **bayesiana**.

Las probabilidades bayesianas son núcleo de varios resultados en teorías de lo más variadas, incluyendo campos de la Economía, Estadística, Filosofía y Física. Se basan en el Teorema de Bayes para poder calcular probabilidades a falta de información observable, mediante el uso de supuestos que aparecen explícitamente en los cálculos. Estos supuestos suelen hacerse sobre la función densidad de los datos, por lo que se parte de la premisa que éstos tienen una densidad $f_X$. Como esta densidad depende de parámetros (como $n$, $p$, $\lambda$, $\mu$, $\sigma$, etc) cuyo valor tampoco conocemos, tenemos que formular el Teorema de Bayes sobre ellos, de forma que con los pocos datos con que contamos podamos ver qué tan probable es el valor de cada parámetro. Se tiene

$$f(\theta|x)=\frac{f(x|\theta)f(\theta)}{f(x)}$$

Una aplicación directa de este teorema está en los paseos aleatorios dados por el algoritmo de Metropolis-Hastings, diseñado para recorrer una función objetivo de forma aleatoria con la intención de explorar sus valores según alguna probabilidad. La función objetivo se denota por $f$, la cual podemos asumir aproximadamente proporcional con alguna función densidad $P(x)$.

Se inicia con un valor cualquiera $x_0$ y con la definición de una función densidad asumida correcta para el valor que podría tener el siguiente valor, $g(x|x_0)$. Por ejemplo, se puede escoger $g(x|x_0)$ como una distribución normal, centrada en $x_0$. Así, los valores más cercanos a $x_0$ son con mayor probabilidad el siguiente paso.

Para el siguiente paso, se muestrea $x'$ de forma aleatoria usando la densidad $g(x|x_0)$, se genera un valor aleatorio $u\sim\mathcal{U}[0,1]$ y se calcula un radio de aceptación, 

$$\alpha=\frac{f(x')}{f(x_0)}=\frac{P(x')}{P(x_0)}$$

Si $u\le\alpha$, se acepta el valor $x'$ como el siguiente paso en la secuencia, $x_1=x'$. Si $u>\alpha$, el valor muestreado se rechaza y se toma $x_1=x_0$, repitiendo iterativamente esta secuencia de pasos hasta un tamaño de muestra $N$ predefinido.

Este algoritmo se puede aplicar a funciones de una o de varias variables, por lo cual es usado en la exploración de espacios de parámetros de forma tal que se puede probar modelos de distinta índole en varias muestras, sin tener que tener una idea preconcebida muy compleja sobre los valores a explorar. Esto da robustez a los modelos y sus resultados.

```python
# Importando las librerías
import matplotlib.colors as mcolors
import matplotlib.pyplot as plt
import numpy as np
import experimento7 as ex
```

```python
def f(x):
    return 2 * x + 1

def Metropolis_iter(x0):
    x_prima = np.random.normal(f(x0), 1)
    u = np.random.uniform() # Python asume u(0,1)
    alpha = f(x_prima) / f(x0)
    
    if alpha <= u:
        x0 = x_prima
    
    return x0
```

```python
x = np.linspace(-5, 15, 300)
y = f(x)
plt.plot(x, y, '-b')
plt.show()
```

```python
x_0 = 0
N = 1000

x_metropolis = []

x_siguiente = Metropolis_iter(x_0)
x_metropolis.append(x_siguiente)

for i in range(1, N):
    x_siguiente = Metropolis_iter(x_siguiente)
    x_metropolis.append(x_siguiente)

x_metropolis = np.asarray(x_metropolis)
y_metropolis = f(x_metropolis)
```

```python
plt.plot(x, y, '-b')
plt.plot(x_metropolis, y_metropolis, 'r')
plt.show()
```

No obstante, la forma más conveniente de ver el camino aleatorio que sigue nuestra distribución construida es tomando el número de veces que aplicamos nuestro método Metrópolis como variable independiente, y graficando la tendencia que sigue el camino realizado.

```python
fig = plt.figure(figsize=(12,5))
plt.plot(y_metropolis, '-r')
plt.show()
```

### Caminos aleatorios en 2D

Aplicaremos ejemplos sencillos de lo anterior para generar caminos aleatorios en espacios de mayor dimensión. Las formas de hacer esto son infinitas y dependen prácticamente de forma exclusiva de quien lo utiliza.

```python
def camino_aleatorio2D(x0, y0, n):
    x, y = x0, y0
    
    xcamino = [x]
    ycamino = [y]
    
    for i in range(n):
        dx, dy = 0, 0
        
        paso = np.random.choice(['E', 'O', 'N', 'S'])
        
        if paso == 'E':
            dx = 1
        elif paso == 'O':
            dx = -1
        elif paso == 'N':
            dy = 1
        else:
            dy = -1
        
        x += dx
        y += dy
        
        xcamino.append(x)
        ycamino.append(y)
    
    xcamino = np.asarray(xcamino)
    ycamino = np.asarray(ycamino)
    
    return xcamino, ycamino
```

```python
x0 = 0
y0 = 0

n = 100000

x, y = camino_aleatorio2D(x0, y0, n)
```

```python
# Usamos colores graduados para conoce dónde comienza el camino (azul) y dónde termina (amarillo)
T = np.linspace(0, 1, np.size(x)) ** 2
s = 10

for i in range(0, n - s, s):
    plt.plot(x[i : i + s + 1], y[i : i + s + 1], color=(T[i], T[i], 1 - T[i]))

plt.show()
```
